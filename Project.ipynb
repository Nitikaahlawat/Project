# **Data Cleaning and Transformation**

## **1. Casualty Dataset**

### **Merge the two columns Seat Belt and Helmet of Casualty Dataset into a new column**

```python
 casu_2019['Safety_Pre_Cas']=casu_2019['Seat Belt'].combine_first(casu_2019['Helmet'])
 casu_2019.info()
```

### **Removing unwanted Columns from the dataset**

```python
casu_2019=casu_2019.drop( 'Seat Belt', axis=1)
casu_2019=casu_2019.drop('Helmet',axis=1)
casu_2019=casu_2019.drop('Hospital',axis=1)
```

### **Removing NAN Values from the dataset**

```python
casu_2019=casu_2019. dropna()
```

## **Renaming some columns**

```python
casu_2019 = casu_2019.rename(columns={'Sex': 'Sex_Casu', 'AGE': 'Age_Casu'})
casu_2019
```

```python
casu_2019.info()
```

## **2. Crash Dataset**

### **Getting information about null values**

```python
crash_2019.isnull().sum()
```

### **Removing some columns**

```python
crash_2019=crash_2019.drop( 'Suburb', axis=1)
crash_2019=crash_2019.drop( 'Postcode', axis=1)
crash_2019=crash_2019.drop( 'LGA Name', axis=1)
crash_2019=crash_2019.drop( 'DUI Involved', axis=1)
crash_2019=crash_2019.drop( 'Drugs Involved', axis=1)
```

```python
crash_2019.info()
```

## **3. Unit Dataset**

### **Getting information about null values**

```python
unit_2019.isnull().sum()
```

### **Removing some columns**

```python
unit_2019=unit_2019.drop( 'Towing', axis=1)
unit_2019=unit_2019.drop( 'Postcode', axis=1)
unit_2019=unit_2019.drop( 'Rollover', axis=1)
unit_2019=unit_2019.drop( 'Fire', axis=1)
```

### **Removing null values from the Dataset**

```python
unit_2019=unit_2019. dropna()
unit_2019.info()
```

## **Merging the two datasets Casualty and Crash**

```python
df1 = pd.merge(casu_2019,crash_2019)
df1
```

# **Merging all the Unit dataset with Casualty and Crash Datasets**

```python
df = pd.merge(df1, unit_2019)
df
```

## **Removing Null values from the dataset**

```python
df= df.dropna()
```

```python
df = df.dropna(how='any',axis=0)
df
```
